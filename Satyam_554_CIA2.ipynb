````python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
````

````python
text = ("Transformer is an exceptional innovation in the field of Deep Learning, contributed by Ashish Vaswani et al. (2017), Google." +
        "The transformer is the most influential Neural Network model that has shown outstanding performance on various NLP tasks including Machine Reading Comprehension, Machine translation and sentence classification." +
        "Attention mechanism and parallelization are the prominent features in the transformers." +
        "Consequently, it can facilitate long-range dependencies without any gradient vanishing or gradient explosion problems and it overcomes the drawbacks of the existing methods such as RNN and LSTM." +
        "The transformer is executed with an encoder-decoder mechanism and the original article of transformers # “Attention All You Need”."
)
````

````python
text
````

````python
word_tokens = nltk.word_tokenize(text)
print (word_tokens)
````

Word Tokenization is used to separate and visualize each work seprately from the assigned paragaraph.

````python
sentence_tokens = nltk.sent_tokenize(text)
print (sentence_tokens)
````

Sentence Tokenization is used to seperate each sentence and visualize it seperately from the assigned paragraph

````python
words = word_tokenize(text)
````

````python
stop_words = set(stopwords.words('english'))
text1 = [word for word in words if word.lower() not in stop_words]
````

Stop Words remove all the less meaningful grammatical words from the paragraph

````python
text1 = ' '.join(text1)
print("Text after removing stopwords:\n", text1)
````

````python
import string
````

````python
punc_count = sum(1 for char in text1 if char in string.punctuation)
text3 = ''.join(char for char in text1 if char not in string.punctuation)
print("Number of punctuation characters:", punc_count)
print("Text without punctuation:", text3)
````

````python
import matplotlib.pyplot as plt
from nltk.probability import FreqDist
````

````python
words1 = word_tokenize(text3)
````

````python
distribution = FreqDist(words1)
top = distribution.most_common(20)
words, frequencies = zip(*top)
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies, color='skyblue')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.title('Frequency Distribution')
plt.xticks(rotation=45, ha='right')
plt.show()
````

This is a visualization using the word tokenization as if we use the paragraph after the metrics of removal of punctuation and stopwords it is identifying only the letters and not the words. Hence multiple visualization is taking place here.

````python
from nltk.stem import PorterStemmer, LancasterStemmer, WordNetLemmatizer
````

Porter STemmer: This stemming technique cuts out the suffix of a word
Lancaster Stemmer: This stemming technique has a set of rules which defines whether a word needs to be deleted or replaced at the end
WOrdNEt LEmmatization: It returns the dictionary value of the word

````python
nltk.download('wordnet')
````

````python
p_stemmer = PorterStemmer()
l_stemmer = LancasterStemmer()
w_lemmatizer = WordNetLemmatizer()
````

````python
SW_P = [p_stemmer.stem(word) for word in words]
SW_L = [l_stemmer.stem(word) for word in words]
L_WOrds = [w_lemmatizer.lemmatize(word) for word in words]
````

````python
print("Porter Stemmer:")
print(SW_P)
print("\nLancaster Stemmer:")
print(SW_L)
print("\nWordNet Lemmatizer:")
print(L_WOrds)
````

````python
nltk.download('averaged_perceptron_tagger')
from collections import Counter
````

````python
pos_tags = nltk.pos_tag(word_tokens)
pos_tags_count = Counter(tag for _, tag in pos_tags)
````

````python
plt.bar(pos_tags_count.keys(), pos_tags_count.values())
plt.xlabel('POS Tags')
plt.ylabel('Frequency')
plt.title('Distribution of POS Tags in Text')
plt.xticks(rotation=45, ha='right')
plt.show()
````
