```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
```

```python
tokens = word_tokenize(text)
```

```python
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.isalpha() and token.lower() not in stop_words]
unique_tokens_count = len(set(filtered_tokens))
print("Number of unique tokens:", unique_tokens_count)
unique_tokens = set(filtered_tokens)
print("Unique Tokens:", unique_tokens)
```

```python
from nltk.probability import FreqDist
```

```python
import re
from datetime import datetime
```

```python
date_pattern = re.compile(r'\b(\d{1,2})[/-](\d{1,2})[/-](\d{4})\b')
matches = date_pattern.findall(text)
```

```python
converted_dates = []
for match in matches:
    day, month, year = map(int, match)
    converted_date = datetime(year, month, day).strftime('%d-%m-%Y')
    converted_dates.append(converted_date)
```

```python
for i in range(len(matches)):
    text = text.replace(matches[i][0], converted_dates[i], 1)
```

```python
print("Original text:")
print(text)
```

```python
year_pattern = re.compile(r'\b\d{4}\b')
years_found = year_pattern.findall(text)
```

```python
year_counts = {}
for year in years_found:
    if year in year_counts:
        year_counts[year] += 1
    else:
        year_counts[year] = 1
```

```python
plt.bar(year_counts.keys(), year_counts.values())
plt.xlabel('Year')
plt.ylabel('Frequency')
plt.title('Distribution of Years in Text')
plt.show()
```

```python
def is_valid_phone_number(phone_number):
    return phone_number.isdigit()
```

```python
def remove_invalid_phone_numbers(text):
    phone_number_pattern = re.compile(r'\b(?:\+\d{1,2}\s?)?[-.\(\)\s\d]+\b')
    phone_numbers_found = phone_number_pattern.findall(text)
    for phone_number in phone_numbers_found:
        if not is_valid_phone_number(''.join(filter(str.isdigit, phone_number))):
            text = text.replace(phone_number, '')

    return text
```

```python
text_without_invalid_phones = remove_invalid_phone_numbers(text)
print("Original Text:")
print(text)
print("\nText without Invalid Phone Numbers:")
print(text_without_invalid_phones)
```
